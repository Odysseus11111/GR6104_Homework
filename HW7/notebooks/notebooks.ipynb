# Create Dataset
We consider a uniform mixture of four two-dimensional Gaussian distributions:

$$
\frac{1}{4}\sum_{k=1}^4 N(\mu_k, \sigma^2 I),
$$

where

$$
\mu_1 = (1,0), \quad
\mu_2 = (-1,0), \quad
\mu_3 = (0,1), \quad
\mu_4 = (0,-1).
$$

In this question, we let $\sigma = 1/4, \delta = 2, n=1000,  \delta * \sigma  =1/2$

```julia
using LinearAlgebra
using Distributions

sigma = 1/4
delta = 2
mix =MixtureModel(
    [
    MvNormal([1.0,0.0],sigma^2*I),
    MvNormal([-1.0,0.0],sigma^2*I),
    MvNormal([0.0,1.0],sigma^2*I),
    MvNormal([0.0,-1.0],sigma^2*I),
    ],
    fill(1/4,4)
)
X = rand(mix,1000)

edge =[]
n = size(X,2)
for i in 1:n
    for j in i+1:n
        dist = norm(X[:,i]-X[:,j])
        if dist <= delta *sigma
            weight = exp(-dist^2/sigma^2)
            push!(edge,(i,j,weight))
        end
    end
    return X, edge
end
```
